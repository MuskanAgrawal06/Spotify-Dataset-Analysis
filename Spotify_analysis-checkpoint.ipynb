```python
select_column= df.sort_values(by= 'Track Score', ascending= False)
selected_column =select_column[['Track', 'Track Score']].head(10)
print(selected_column) #filtering out the top 10 tracks on the basis of their track score
```

```python
plt.figure(figsize=(25,10))
bar_width=1
plt.bar(selected_column['Track'],selected_column['Track Score'])
plt.xlabel('Track Name')
plt.ylabel('Track Score')
plt.tight_layout()
plt.show() #visualizing the track and track score
```

```python
#comparison between spotify playlist count and azamzon playlist count
sorted= df.sort_values(by='Spotify Playlist Count',ascending= False)
plt.figure(figsize=(50,20))
bar_width=2
positions=np.arange(len(sorted))

spotify_bars=plt.bar(sorted['Track'],sorted['Spotify Playlist Count'],label= 'Spotify', color='blue')

apple_bars=plt.bar(sorted['Track'],sorted['Apple Music Playlist Count'],label= 'apple', color='coral')
plt.legend()
plt.show()

```

```python
spotifyname= df.nlargest(10,'Spotify Popularity')
print(spotifyname)
plt.figure(figsize=(8, 8))
plt.pie(spotifyname['Spotify Popularity'],labels=spotifyname['Track'], autopct='%1.1f%%')
plt.title('Top 5 Genres on Spotify')
plt.show()
```

```python
# Get the top 10 tracks by Spotify popularity
df= pd.read_csv('Most Streamed Spotify Songs 2024.csv',encoding= 'unicode_escape')
top_tracks = df.nlargest(10,'Apple Music Playlist Count')


# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_tracks['Apple Music Playlist Count'], labels=top_tracks['Track'], autopct='%1.1f%%')
plt.title('Top 10 Tracks on Spotify by Popularity')
plt.show()
```
